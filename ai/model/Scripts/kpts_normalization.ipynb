### keypoint normalization function

```python
def kpts_change(kpts):
    kpts = [round(float(x), 2) for x in kpts]
    kpts = np.array(kpts)
    kpts[0::3] /= img_size_w
    kpts[1::3] /= img_size_h
    return kpts
```
